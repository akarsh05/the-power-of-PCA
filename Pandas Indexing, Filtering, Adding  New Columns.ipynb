```python
## Keep in mind
# iloc - index based selection. Index position, not its actual value, matters.
# loc - label based selection. Index value, not position, matters.
# differenc between iloc and loc
# loc incex includes last element as well
# iloc [0:10] -> 9 elements, loc[0:10] -> 10 elements
# for loc, this is benefitial when index is strings. Easier to define range as last element is included.
# both loc and iloc are row-first, column-second, opposite of native python
# can use lists, negative values for index

## adding new colums to a DF
# insert - add column to an existing dataframe
# assign - create a new dataframe with the new column
```
